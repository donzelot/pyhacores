```python
fft_size = 1024 * 8
```

```python
hann_fix9 = np.array([Sfix(x, 0, -6, round_style='round', overflow_style='saturate').val for x in  np.hanning(fft_size)])
```

```python
sig = np.exp(2j*np.pi*np.linspace(0, 1, fft_size) * 100) * 0.125
sig = sig * hann_fix9

sig2 = np.exp(2j*np.pi*np.linspace(0, 1, fft_size) * 110) * 0.125
sig2 = sig2 * hann_fix9

sig = sig + sig2
```

```python
plt.plot(sig)
plt.show()
```

```python
dut = R2SDF(fft_size)
sims = simulate(dut, sig, simulations=['PYHA'], output_callback=unpackage, input_callback=package)
```

```python
def bit_reverse(x, n_bits):
    return int(np.binary_repr(x, n_bits)[::-1], 2)

def bit_reversed_indexes(N):
    return [bit_reverse(i, int(np.log2(N))) for i in range(N)]

pyh = np.array(sims['PYHA'][0])
pyh = pyh[bit_reversed_indexes(fft_size)]
pyh = np.fft.fftshift(pyh)
pyh = np.abs(pyh)
pyh = pyh / pyh.max()
```

```python
a = 20 * np.log10(pyh)
plt.plot(a)
# plt.magnitude_spectrum(sig, window=matplotlib.mlab.window_none, scale='dB')
plt.show()
```

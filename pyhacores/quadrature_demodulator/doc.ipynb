```python
import numpy as np
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
# %matplotlib inline
%load_ext autoreload
%autoreload 2

from pyhacores.filter.fir.model import FIR
from pyhacores.quadrature_demodulator.model import QuadratureDemodulator
from scipy import signal
from pyha.common.util import plot_freqz
```

```python
close = np.load('data/taranis_sps16_fs1600000.0_band1000000.0_far.npy')
# close = close[:len(close)//4]


taps = signal.remez(64, [0, 0.05, 0.15, 0.5], [1, 0])
plot_freqz(taps)
dut = FIR(taps)
y = dut.model_main(close)

plt.plot(abs(y))
plt.show()


demod = QuadratureDemodulator()
y = demod.model_main(y[840000:860000])

plt.plot(y)
plt.show()
```
